### Importing Library

```python
import numpy as np
```

### Hebbs Rule

```python
# Define the input vectors
xl = np.array([1, 0, 1])
x2 = np.array([0, 1, 0])

# Initialize the weight matrix
w = np.zeros((3, 3))

# Apply Hebb 's rule to update the weight matrix
w += np.outer(xl, xl)
w += np.outer(x2, x2)

print("Weight matrix after applying Hebb's rule: ")
print(w)
```

### Delta Rule / Widrow-Hoff Rule

```python
# Define the input vector and desired output
x = np.array([1, 0, 1])
d = 1

# Initialize the weight vector and leanring rate
w = np.zeros(3)
eta = 0.1

# Repeat the training process until convergence
for i in range(100):
    # Compute the output
    y = np.dot(w, x)
    
    # Comput ethe erro between the network's and the desired output
    e = d - y
    
    # Update the weight vector using Delta Rule
    w += eta * e * x
    
    # Print the error and the weight vector
    print("Iteration: ", i + 1)
    print("Error: ", e)
    print("Weight Vector:", w, "\n")
```
